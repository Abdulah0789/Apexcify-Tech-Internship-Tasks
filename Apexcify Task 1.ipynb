```python
import random

print("Welcome to Rock-Paper-Scissors!")
print("Type 'quit' to stop playing.\n")

options = ["rock", "paper", "scissors"]

while True:
    user_input = input("Enter Rock, Paper, or Scissors: ").lower()

    if user_input == "quit":
        print("Thanks for playing! Goodbye.")
        break

    if user_input not in options:
        print("Invalid input. Please check your spelling and try again.")
        continue

    computer_pick = random.choice(options)
    print(f"Computer chose: {computer_pick}")

    if user_input == computer_pick:
        print("It's a tie! ðŸ‘”")
    
    elif (user_input == "rock" and computer_pick == "scissors") or \
         (user_input == "paper" and computer_pick == "rock") or \
         (user_input == "scissors" and computer_pick == "paper"):
        print("You win! ðŸŽ‰")
    
    else:
        print("You lose! Computer wins this round. ðŸ¤–")
    
    print("-" * 30)
```
